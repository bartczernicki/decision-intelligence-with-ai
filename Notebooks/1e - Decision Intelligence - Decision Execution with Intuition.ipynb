### Introducing Decision Execution with Intuition 

<img style="display: block; margin: auto;" width ="700px" src="../Images/DecisionIntelligence-DecisionExecution.png">

> 📜 **_"All of my best decisions in business and in life have been made with heart, intuition, guts... not analysis. 
> When you can make a decision with analysis, you should do so. But it turns out, in life, your most important
> decisions are always made with instinct, intuition, taste, heart."_**
>
> -- <cite>Jeff Bezos (founder of Amazon, Space Entrepreneur)</cite>  

The above quote from Jeff Bezos above underscores the balance between data-driven decision-making and the role of intuition and instinct in navigating complex or uncertain situations. It is important to start with intuition, as it is how humans learn to make decisions. As young infants, we have learned what decisions to make based on intuition. Intuition is how most of us have executed most decisions throughout life.  

**How is intuition defined? Intuition is best described as the ability to understand something immediately, without the need for conscious reasoning.** Others describe it in more simpler terms as a sixth-sense or a "gut feel". It's based on identifying patterns and experiences that the brain has internalized over time. Intuitive judgments can be rapid and efficient, drawing on a wealth of experience & knowledge. Therefore, intuition can be a highly effective decision-making tool by people who are highly experienced. For example, a basketball coach for 20 years has probably experienced similar patterns in basketball game scenarios allowing the coach to make immediate decisions for his team impacting a game in real-time. Another example is driving a car or riding a bike. When you are driving a car, you are not doing complex propulsion nor physics calculations in your head. The driving process is simplified to you from all the past driving experiences where it becomes second nature. In casual driving in familiar surroundings, you can almost "turn off" focusing on driving decisions and think about other priorities. Intuition is a highly effective ability when decision-making needs to be as quick as possible and you have the experience to **recognize patterns** very quickly. 

To define intuition another way, below are two quotes from Daniel Kahneman:

> 📜 **_"Intuition is thinking that you know without knowing why you do."_**
>
> 📜 **_"Intuition is nothing more and nothing less than recognition."_**
> 
> -- <cite>Daniel Kahneman (Nobel laureate renowned for his groundbreaking work on the psychology of judgment, decision-making, and behavioral economics)</cite>  

### Intuition - Scenario: A Firefighter's Split-Second Decision Making

<img style="float: left;padding-right: 25px" width ="400px" src="../Images/DecisionIntelligenceFramework/DecisionExecution-Intuition-Firefighter.png">

Imagine a scenario where a firefighter responds to a 3-alarm fire at a multi-story house. As he arrives on the scene, he sees flames engulfing the top floor and people trapped inside.  He analyzes the situation: The building's layout is unfamiliar. The fire is spreading rapidly, and the stairwell is impassable. There are only two minutes to rescue those trapped before the floor collapses.  

In this situation, years of training and intuition takes over. The firefighter recalls a similar incident where a hidden entrance was discovered on the opposite side of the building. Trusting an internal gut feeling, the firefighter decides to lead a team around the building to search for an alternate entry point. In this high-pressure situation, the firefighter's intuition proves more valuable than following standard decision models, allowing him to save more lives. 

Scenarios like this demonstrate the value of Intuition applied Decision Making:  
- Time is of the essence  
- Information is limited or uncertain (unknown house layout)  
- Experience and pattern recognition are essential (training comes out naturally without pauses for reflection)

### Intuition - When We Need More Powerful Execution Methods 

Using intuition is the most common decision execution framework. However, it is not well suited for all decisions. When is intuition (recognition) NOT a great tool for decision-making? 
* The decision is sufficiently important (buying a house, committing to a four year University)
* When one is presented with a great deal of supporting information to be analyzed 
* There is ample time to improve the quality of the decision (no need to rush)
* When the situation is "new", but there is enough historical decision outcomes  
* When the human making a mathematic or probabilistic decision is not calibrated (if you are presented the same decision scenario, you can decide many different ways)

In the decision situations listed above, intuition can be unreliable. Therefore, there are better decision execution options than intuition if you have the time, the data or can draw on past repeatable decision outcomes.  

### Intuition - Generative AI 

<img style="display: block; margin: auto;" width ="600px" src="../Images/DecisionIntelligence-DecisionExecution-DescartesQuote.png">

What does intuition have to do with Generative AI? Most correlate intuition with human cognitive behavior and currently we don't consider Generative AI LLMs as sentient beings. It turns out intuition decision-making techniques can be applied to Generative AI. In the context of Generative AI, **intuition is basically allowing the Generative AI model make a decision from it's own knowledge & patterns recognition it has learned from the mountains of training data**. In the same way that intuition is a rapid reasoning framework to make the decisions for humans, it is the same for Generative AI. This means that when you ask a logic question, reasoning, recommendation or for Generative AI to make a potential decision it will respond much faster following the straightforward path while activating neurons in its path. You may trust a GenAI's "intuition" rapid response system to recommend simple scenarios, but not for making high-stakes decisions without specific instructions, knowledge or other guardrails. 

<img style="float: left;padding-right: 25px" width ="400px" src="../Images/DecisionIntelligenceFramework/DecisionExecution-Intuition-BabyLearning.png">

There is an interesting connection in how humans learn from childhood to adulthood and how GenAI models are trained with data to develop their own "AI intuition". Each human has their own personal life examples they are presented during their formative period or adult career. These life examples vary in the specificity of the type of examples (data), quality of examples (professional teaching or self-taught), the amount of examples and finally how we as humans internalize them. The same applies for Generative AI! When Generative AI models are trained, they are presented with varying quality and quantity data sets. They are instructed to "learn" from the presented data using varying algorithmic and mathematic approaches. Finally, they are trained using varying durations of time. The outcome from this training is a model that recognizes patterns and communicates them back to the user. What is more interesting as you vary the data, type of training, duration or training or size of the model's memory you get vastly different GenAI capabilities. Just like people vary in their expertise, we have trained GenAI models that differ in their pattern recognition and expertise. For example, GenAI models can be very good at a specific thing (chess), good all-around knowledge experts (Ken Jennings - Jeopardy) or good at selected  domains (coding, math, finance) etc.

If you send a basic prompt to a GenAI model you are entrusting the GenAI's intuition (it's knowledge, pattern recognition and experience it has learned). Therefore, the more basic the prompt and less specific the decision instructions, the more you are entrusting GenAI's intuition, recognition and learned pattern discovery. Unfortunately, this is where most decision-makers start and terminate their journey relying on AI's "intuition" alone. This isn't necessarily bad, but decision-making with AI can be further optimized with more advanced decision execution tools.
